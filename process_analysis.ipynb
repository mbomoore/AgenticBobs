#  Lets make a little framework for analyzing the various different ways we can impact this process


Some data:

Each task takes 30 minutes of processing time, every time you start a task which is unrelated to your previous task, there is a 10-minute context switching penalty.

Input tasks arrive at a rate of 1 every 10 minutes.


```python
switching_cost = 10
task1_time = 30
task2_time = 30
task3_time = 30

arrival_rate = 1/10
```

# Scenario 1: Each task is done by a different worker from a different resource pool


```python
# Scenario 1 (sim_dsl): Each task is done by a different worker (switching before each task)
from sim_dsl.core import ProcessModel, State
from sim_dsl.resources import Resource

# parameters (minutes)
switching_cost = 10
task1_time = 30
task2_time = 30
task3_time = 30
arrival_rate_per_min = 1/10.0

# convert to hours for cost calculations, but keep minutes for the "processing_time" printouts
s_hr = switching_cost/60.0
t1_hr = task1_time/60.0
t2_hr = task2_time/60.0
t3_hr = task3_time/60.0

p1 = ProcessModel('scenario1_diff_workers')
with p1:
    # three distinct human resources
    h1 = Resource('Worker1', cost=50.0)
    h2 = Resource('Worker2', cost=50.0)
    h3 = Resource('Worker3', cost=50.0)

    start = State('Start', time=0.0)
    sw1 = State('Switch1', time=s_hr)
    task1 = State('Task1', time=t1_hr, resource=h1)
    sw2 = State('Switch2', time=s_hr)
    task2 = State('Task2', time=t2_hr, resource=h2)
    sw3 = State('Switch3', time=s_hr)
    task3 = State('Task3', time=t3_hr, resource=h3)
    success = State('Success')

    # deterministic flow
    p1.add_transition(start >> 1.0 >> sw1)
    p1.add_transition(sw1 >> 1.0 >> task1)
    p1.add_transition(task1 >> 1.0 >> sw2)
    p1.add_transition(sw2 >> 1.0 >> task2)
    p1.add_transition(task2 >> 1.0 >> sw3)
    p1.add_transition(sw3 >> 1.0 >> task3)
    p1.add_transition(task3 >> 1.0 >> success)

# expected visits (relative-to-success normalization inside state_visits)
visits = p1.state_visits(starting_state=0)
# total processing time per case (hours -> minutes)
total_time_hr = sum(s.time * v for s, v in visits)
total_time_min = total_time_hr * 60.0

print(f"Scenario 1 total processing time: {total_time_min:.1f} minutes")
print(f"Resources needed (Little's law, arrival rate={arrival_rate_per_min}/min): {total_time_min * arrival_rate_per_min:.3f}")

# estimated cost per case (resources cost in $/hr)
cost_per_case = 0.0
for s, v in visits:
    if getattr(s, 'resource', None):
        cost_per_case += s.resource.cost * s.time * v
print(f"Estimated cost per case: ${cost_per_case:.2f}")
```

```python
# little's law
print(resources_needed := total_time_min * arrival_rate_per_min)
```

# Scenario 2: Obviously it would be better to do this with a single person



```python
# Scenario 2 (sim_dsl): Single person does all three tasks (one switching before work)
from sim_dsl.core import ProcessModel, State
from sim_dsl.resources import Resource

# parameters (minutes)
switching_cost = 10
task1_time = 30
task2_time = 30
task3_time = 30
arrival_rate_per_min = 1/10.0

s_hr = switching_cost/60.0
t1_hr = task1_time/60.0
t2_hr = task2_time/60.0
t3_hr = task3_time/60.0

p2 = ProcessModel('scenario2_single_worker')
with p2:
    # single human resource
    worker = Resource('SoloWorker', cost=50.0)

    start = State('Start', time=0.0)
    switch = State('Switch', time=s_hr)
    task1 = State('Task1', time=t1_hr, resource=worker)
    task2 = State('Task2', time=t2_hr, resource=worker)
    task3 = State('Task3', time=t3_hr, resource=worker)
    success = State('Success')

    p2.add_transition(start >> 1.0 >> switch)
    p2.add_transition(switch >> 1.0 >> task1)
    p2.add_transition(task1 >> 1.0 >> task2)
    p2.add_transition(task2 >> 1.0 >> task3)
    p2.add_transition(task3 >> 1.0 >> success)

visits = p2.state_visits(starting_state=0)
total_time_hr = sum(s.time * v for s, v in visits)
total_time_min = total_time_hr * 60.0

print(f"Scenario 2 total processing time: {total_time_min:.1f} minutes")
print(f"Resources needed (Little's law): {total_time_min * arrival_rate_per_min:.3f}")

cost_per_case = sum((s.resource.cost * s.time * v) if getattr(s, 'resource', None) else 0.0 for s, v in visits)
print(f"Estimated cost per case: ${cost_per_case:.2f}")
```

```python
# little's law
print(resources_needed := total_time_min * arrival_rate_per_min)
```

# Scenario 3: AgenticAI Does Middle Step


```python
# Scenario 3 (sim_dsl): AI does the middle step; on AI failure escalate to human (incurs switching)
from sim_dsl.core import ProcessModel, State
from sim_dsl.resources import Resource

# parameters (minutes)
switching_cost = 10
task1_time = 30
task2_time = 30
task3_time = 30
agent_processing_min = 1
agent_success_rate = 0.8
arrival_rate_per_min = 1/10.0

s_hr = switching_cost/60.0
t1_hr = task1_time/60.0
t2_hr = task2_time/60.0
t3_hr = task3_time/60.0
agent_hr = agent_processing_min/60.0

p3 = ProcessModel('scenario3_agent_middle')
with p3:
    human = Resource('Human', cost=50.0)
    ai = Resource('AI', cost=5.0)

    start = State('Start', time=0.0)
    switch1 = State('SwitchToHuman1', time=s_hr)
    task1 = State('Task1', time=t1_hr, resource=human)

    switch_to_ai = State('SwitchToAI', time=0.0)  # assume no switching cost to call AI
    task2_ai = State('Task2_AI', time=agent_hr, resource=ai)

    # path when AI fails: switching cost to human + human does Task2
    switch_to_human = State('SwitchToHumanAfterFail', time=s_hr)
    task2_human = State('Task2_Human', time=t2_hr, resource=human)

    switch_back = State('SwitchToHumanForTask3', time=s_hr)
    task3 = State('Task3', time=t3_hr, resource=human)
    success = State('Success')

    # flow
    p3.add_transition(start >> 1.0 >> switch1)
    p3.add_transition(switch1 >> 1.0 >> task1)
    p3.add_transition(task1 >> 1.0 >> switch_to_ai)
    p3.add_transition(switch_to_ai >> 1.0 >> task2_ai)

    # AI outcomes
    p3.add_transition(task2_ai >> agent_success_rate >> switch_back)
    p3.add_transition(task2_ai >> (1-agent_success_rate) >> switch_to_human)

    # failure escalates to human do Task2 then proceed
    p3.add_transition(switch_to_human >> 1.0 >> task2_human)
    p3.add_transition(task2_human >> 1.0 >> switch_back)

    # after success or after human rework, go to task3
    p3.add_transition(switch_back >> 1.0 >> task3)
    p3.add_transition(task3 >> 1.0 >> success)

visits = p3.state_visits(starting_state=0)
total_time_hr = sum(s.time * v for s, v in visits)
total_time_min = total_time_hr * 60.0

print(f"Scenario 3 total processing time (mean per case): {total_time_min:.1f} minutes")
print(f"Resources needed (Little's law): {total_time_min * arrival_rate_per_min:.3f}")

cost_per_case = sum((s.resource.cost * s.time * v) if getattr(s, 'resource', None) else 0.0 for s, v in visits)
print(f"Estimated cost per case: ${cost_per_case:.2f}")
```

Some stuff to notice. 

- We did not save 29 minutes of total time, but only 11 total minutes. 10 minutes where lost to additional context switching for the person who now has to come back in to do task 3. We might think about going back to the worker who did task 1 to reduce context switching costs, but that put additional constraints on our resource pool. In addition, 20% of the time we have to have the user do it again, which adds .2*(40) to the process.

- Well, you might say, we could just have the AI do more of the process, then humans won't have to context switch as much. But, now you have a new problem, which is uncaught errors. If the AI does 2 parts of the process with no human checking, the chance that it has messed up by the end is the 1-(p_success_1 * p_success_2), creating compounding errors.



# 4 AI does the last two steps, human check at the end

(the human check requires the context switching tax, and on fail, the actual work time.)


```python
# Scenario 4 (sim_dsl): AI does last two steps; human check at end (human check incurs switching cost)
from sim_dsl.core import ProcessModel, State
from sim_dsl.resources import Resource

# parameters (minutes)
switching_cost = 10
task1_time = 30
task2_time = 30
task3_time = 30
agent_processing_min = 1
agent_success_rate = 0.8
arrival_rate_per_min = 1/10.0

s_hr = switching_cost/60.0
t1_hr = task1_time/60.0
t2_hr = task2_time/60.0
t3_hr = task3_time/60.0
agent_hr = agent_processing_min/60.0

p4 = ProcessModel('scenario4_ai_last_two')
with p4:
    human = Resource('Human', cost=50.0)
    ai = Resource('AI', cost=5.0)

    start = State('Start', time=0.0)
    switch1 = State('SwitchToHuman1', time=s_hr)
    task1 = State('Task1', time=t1_hr, resource=human)

    # AI handles task2 and task3
    task2_ai = State('Task2_AI', time=agent_hr, resource=ai)
    task3_ai = State('Task3_AI', time=agent_hr, resource=ai)

    # human check (incurs switching)
    switch_check = State('SwitchToHumanForCheck', time=s_hr)
    human_check = State('HumanCheck', time=0.0, resource=human)

    # on check failure -> human rework does task2+task3 (assume done by one human sequentially)
    rework = State('Rework_Human', time=(t2_hr + t3_hr), resource=human)
    success = State('Success')

    # flow
    p4.add_transition(start >> 1.0 >> switch1)
    p4.add_transition(switch1 >> 1.0 >> task1)
    p4.add_transition(task1 >> 1.0 >> task2_ai)
    p4.add_transition(task2_ai >> 1.0 >> task3_ai)
    p4.add_transition(task3_ai >> 1.0 >> switch_check)

    # check outcomes
    p4.add_transition(switch_check >> 1.0 >> human_check)
    overall_success_prob = agent_success_rate * agent_success_rate
    p4.add_transition(human_check >> overall_success_prob >> success)
    p4.add_transition(human_check >> (1-overall_success_prob) >> rework)
    p4.add_transition(rework >> 1.0 >> success)

visits = p4.state_visits(starting_state=0)
total_time_hr = sum(s.time * v for s, v in visits)
total_time_min = total_time_hr * 60.0

print(f"Scenario 4 total processing time (mean per case): {total_time_min:.1f} minutes")
print(f"Resources needed (Little's law): {total_time_min * arrival_rate_per_min:.3f}")

cost_per_case = sum((s.resource.cost * s.time * v) if getattr(s, 'resource', None) else 0.0 for s, v in visits)
print(f"Estimated cost per case: ${cost_per_case:.2f}")
```

# Scenario 5: AI does all three tasks with a human check at the end

```python
# Scenario 5 (sim_dsl): AI does all three tasks; human check at end, possible full rework
from sim_dsl.core import ProcessModel, State
from sim_dsl.resources import Resource

# parameters (minutes)
switching_cost = 10
task1_time = 30
task2_time = 30
task3_time = 30
agent_processing_min = 1
agent_success_rate = 0.8
arrival_rate_per_min = 1/10.0

s_hr = switching_cost/60.0
t1_hr = task1_time/60.0
t2_hr = task2_time/60.0
t3_hr = task3_time/60.0
agent_hr = agent_processing_min/60.0

p5 = ProcessModel('scenario5_ai_all')
with p5:
    human = Resource('Human', cost=50.0)
    ai = Resource('AI', cost=5.0)

    start = State('Start', time=0.0)
    # AI does Task1, Task2, Task3
    task1_ai = State('Task1_AI', time=agent_hr, resource=ai)
    task2_ai = State('Task2_AI', time=agent_hr, resource=ai)
    task3_ai = State('Task3_AI', time=agent_hr, resource=ai)

    # human check (incurs switching)
    switch_check = State('SwitchToHumanForCheck', time=s_hr)
    human_check = State('HumanCheck', time=0.0, resource=human)

    # on check failure -> full human rework (task1+task2+task3)
    rework = State('FullRework_Human', time=(t1_hr + t2_hr + t3_hr), resource=human)
    success = State('Success')

    p5.add_transition(start >> 1.0 >> task1_ai)
    p5.add_transition(task1_ai >> 1.0 >> task2_ai)
    p5.add_transition(task2_ai >> 1.0 >> task3_ai)
    p5.add_transition(task3_ai >> 1.0 >> switch_check)
    p5.add_transition(switch_check >> 1.0 >> human_check)

    overall_success_prob = agent_success_rate ** 3
    p5.add_transition(human_check >> overall_success_prob >> success)
    p5.add_transition(human_check >> (1-overall_success_prob) >> rework)
    p5.add_transition(rework >> 1.0 >> success)

visits = p5.state_visits(starting_state=0)
total_time_hr = sum(s.time * v for s, v in visits)
total_time_min = total_time_hr * 60.0

print(f"Scenario 5 total processing time (mean per case): {total_time_min:.1f} minutes")
print(f"Resources needed (Little's law): {total_time_min * arrival_rate_per_min:.3f}")

cost_per_case = sum((s.resource.cost * s.time * v) if getattr(s, 'resource', None) else 0.0 for s, v in visits)
print(f"Estimated cost per case: ${cost_per_case:.2f}")
```

# Using sim_dsl inside this notebook

Below are short, self-contained examples showing how to build a small `ProcessModel` using `sim_dsl`, inspect its transition matrix and expected state visits, compute simple costs, and run a few stochastic simulation runs with the SimPy adapter.

```python
# Build a small ProcessModel using sim_dsl
from sim_dsl.core import ProcessModel, State
from sim_dsl.resources import Resource

p = ProcessModel('three_step')
with p:
    # resources
    human = Resource('Human', cost=50.0)
    ai = Resource('AI', cost=5.0)

    # states (time in hours for cost calcs; 0.5 hr = 30 minutes)
    s1 = State('Task1', time=0.5, resource=human)
    s2 = State('Task2', time=0.0166667, resource=ai)  # 1 minute = 1/60 hr
    s3 = State('Task3', time=0.5, resource=human)
    success = State('Success')

    # transitions: Task1 always -> Task2; Task2 -> Success with p=0.8 else back to Task3 (escalate)
    p.add_transition(s1 >> 1.0 >> s2)
    p.add_transition(s2 >> 0.8 >> success)
    p.add_transition(s2 >> 0.2 >> s3)
    # Task3 -> Success
    p.add_transition(s3 >> 1.0 >> success)

# show transition matrix
import numpy as np
print('States:', [s.name for s in p.states])
print('Transition matrix:\n', p.transition_matrix())

# compute expected visits starting from Task1
visits = p.state_visits(starting_state=0)
print('\nExpected visits (relative to Success):')
for state, v in visits:
    print(f"  {state.name}: {v:.4f}")
```

```python
# Empirical SimPy runs for scenarios 1..5
# Requires that p1, p2, p3, p4, p5 are defined in the notebook's kernel (they're created above).
from sim_dsl.simpy_adapter import simulate_with_simpy
import numpy as np

models = [
    ('scenario1', p1),
    ('scenario2', p2),
    ('scenario3', p3),
    ('scenario4', p4),
    ('scenario5', p5),
]
runs = 200
seed = 20250819

print(f"Running {runs} runs per scenario (seed={seed})...\n")
for name, model in models:
    res = simulate_with_simpy(model, runs=runs, seed=seed)
    events = res['events']
    counts = res.get('counts', {})
    # gather first Success time per run
    success_times = {}
    for t, run_id, stname in events:
        if stname == 'Success' and run_id not in success_times:
            success_times[run_id] = t
    times_min = [v * 60.0 for v in success_times.values()]
    mean_time = float(np.mean(times_min)) if len(times_min) else None
    success_rate = len(success_times) / runs
    mean_time_str = f"{mean_time:.2f}" if mean_time is not None else "N/A"
    print(f"{name}: success_rate={success_rate:.3f}, mean_time_min={mean_time_str}")
    print(f"  counts sample: {dict(list(counts.items())[:6])}\n")
```
